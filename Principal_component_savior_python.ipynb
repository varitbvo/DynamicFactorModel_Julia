```python
import sklearn.decomposition as decom
import numpy as np
import pandas as pd
```

```python
A = pd.read_csv("prior_matrix.csv", header = None)
#A = np.array([[1,2,-3], [1,-4,3]])
```

```python
pca = decom.PCA(n_components = 118)
```

```python
pca.fit(A)
```

```python
print(pca.explained_variance_ratio_)
```
